```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from utilies.bs_esp import loaddiff2a4z,trimdata
from matplotlib.ticker import ScalarFormatter, LogLocator
```

```python
fsmall = 'small/ESP_RRMSD.txt'
# for small mol, taking CCSD as the reference
METS_small = ['HF','B3LYP','M06','M062X','MN15','WB97XD','MP2']
```

```python
flarge = 'large/ESP_RRMSD.txt'
# for small mol, taking CCSD as the reference
METS_large = ['HF','B3LYP','M06','M062X','MN15','WB97XD']
```

```python
output_s = {}
for smet in METS_small:
    output_s[smet] = loaddiff2a4z(fsmall,smet,column=13)
```

```python
output_l = {}
for lmet in METS_large:
    output_l[lmet] = loaddiff2a4z(flarge,lmet,column=13)
```

```python
xl = ['HF/a5z','B3LYP/a5z','M06/a5z','M06-2X/a5z','MN15/a5z','$\omega$B97X-D/a5z','MP2/a5z']

fig, ax = plt.subplots(figsize=(3,3))

ax1 = ax.boxplot([trimdata(output_s['HF']),trimdata(output_s['B3LYP']),trimdata(output_s['M06']),\
                  trimdata(output_s['M062X']),trimdata(output_s['MN15']),trimdata(output_s['WB97XD']),\
                  trimdata(output_s['MP2'])],\
                    boxprops={'color':'black'},\
                    flierprops={'marker':'*','markeredgecolor':'black'},\
                    whiskerprops={'linestyle':'-','color':'black'},\
                    widths=0.5, positions=np.arange(1,len(METS_small)+1,1))

#means = [np.mean(trimdata(output[col])) for col in ['HF','B3LYP','M06','M062X','MN15','WB97XD']]
#ax.scatter(np.arange(1,len(METS)+1,1),means,color='none',marker='o',edgecolor='black',s=15)

ax.legend([ax1['boxes'][0]],['CCSD/a5z'],ncol=1,loc='upper right')

ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

ax.set_xticklabels(xl,rotation=90,fontsize=12)

#ax.set_ylim(0,0.2)
#ax.yaxis.set_ticks(np.arange(0,0.21,0.05))

plt.grid(ls='--')

#plt.savefig('normal_plot_atzvsmp2a4z.png',dpi=600, bbox_inches='tight')
```
